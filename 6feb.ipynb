```python
'''Q1. Create a function which will take a list as an argument and return the product of all the numbers
after creating a flat list.
Use the below-given list as an argument for your function.
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']
Note: you must extract numeric keys and values of the dictionary also.'''

def flatten_and_product(input_list):
    flat_list = []

    def flatten(lst):
        for item in lst:
            if isinstance(item, (list, tuple, set)):
                flatten(item)
            elif isinstance(item, dict):
                flatten(item.values())
                flatten(item.keys())
            else:
                flat_list.append(item)

    flatten(input_list)

    # Calculate the product of all numeric elements
    product = 1
    for item in flat_list:
        if isinstance(item, (int, float)):
            product *= item

    return product

# Given list
list1 = [1, 2, 3, 4, [44, 55, 66, True], False, (34, 56, 78, 89, 34), {1, 2, 3, 3, 2, 1}, {1: 34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']

# Call the function
result = flatten_and_product(list1)

# Print the result
print("Product of all numeric elements in the flattened list:", result)






#2)

def encrypt_message(message):
    encrypted_text = ""

    for char in message:
        if char.isalpha():
            # Encrypt alphabets
            encrypted_char = chr(ord('z') - (ord(char.lower()) - ord('a')))
            encrypted_text += encrypted_char
        elif char.isspace():
            # Replace whitespace with a dollar sign
            encrypted_text += '$'
        else:
            # Keep punctuation marks unchanged
            encrypted_text += char

    return encrypted_text

# Input sentence
input_sentence = "I want to become a Data Scientist."

# Convert the input sentence to lowercase
lowercase_sentence = input_sentence.lower()

# Encrypt the sentence
encrypted_result = encrypt_message(lowercase_sentence)

# Print the encrypted result
print("Encrypted Message:", encrypted_result)









```
